```python
#import ast
import json
import scrapy
from scrapy.crawler import CrawlerProcess
from bs4 import BeautifulSoup
```

```python
field_list = ['unidade', 'endereco', 'bairro', 'cep', 'sigla', 'funcionamento', 'url']
#field_list = ['unidade', 'endereco', 'bairro', 'treino_1', 'treino_3', 'treino_5', 'treino_10']
```

```python
class BioRitmoItem(scrapy.Item):
    fields = {field: scrapy.Field() for field in field_list}
```

```python
class BioRitmoSpider(scrapy.Spider):
    name = 'bioritmo'
    start_urls = ['https://www.bioritmo.com.br/unidades']

    def parse_location(self, location):
        item = {}
        item['unidade'] = location['name']
        item['sigla'] = location['acronym']
        item['url'] = BioRitmoSpider.start_urls[0] + '/' + location['slug']
        item['funcionamento'] = location['schedule']
        #print('>>>>>>>>>>>>>>>>>>>> ', repr(location['schedule']))
        item['endereco'] = location['address']['street']
        item['bairro'] = location['address']['neighborhood']
        item['cep'] = location['address']['zip']
        return item
    
    def parse(self, response):
        html = BeautifulSoup(response.text)
        props = html.select_one('div[data-react-class="Locations/Index"]').attrs['data-react-props']
        props.replace('null','None')
        locations = json.loads(props)['locations']
        items = [BioRitmoItem(self.parse_location(location)) for location in locations]
        for location in locations:
            #print(location['address']['city_id'])
            for k,v in location.items():
                print(k, v)
            print('--------------------------------------------------------------------------\n')
        #print(items)
```

```python
process = CrawlerProcess()
process.crawl(BioRitmoSpider)
process.start()
```
