```python
class WHOSpider(scrapy.Spider):
    name = 'who-countries'
    allowed_domains = 'www.who.int'
    base_url = 'https://www.who.int'
    start_urls = [base_url + '/countries/en']
    
    def parse(self, response):
        urls = response.css('ul.index > li > a::attr(href)').extract()
        for url in urls:
            yield scrapy.Request(url=self.base_url+url, callback=self.parse_stats, dont_filter=True)

    def parse_stats(self, response):
        global items, field_map
        name = response.css('h1::text').extract_first()
        rows = response.css('table.tableData > tbody > tr')
        
        item = {}
        item['Name'] = name
        
        for row in rows:
            field = row.css('th::text').extract_first()
            value = row.css('td > strong::text').extract_first()
            item[field] = value

        items.append(item)
```

```python
process = scrapy.crawler.CrawlerProcess()
process.crawl(WHOSpider)
process.start()
```
